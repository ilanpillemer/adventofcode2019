````julia
mutable struct Moon
    px
    py
    pz
    vx 
    vy
    vz
end
````

````julia
function dowork(moons)
    #To apply gravity, consider every pair of moons. 
    #On each axis (x, y, and z), the velocity of each moon changes by exactly +1 or -1 
    #to pull the moons together. 
    
    #For example, if Ganymede has an x position of 3, 
    #            and Callisto has a x position of 5, 
    #then Ganymede's x velocity changes by +1 (because 5 > 3) and 
    # Callisto's x velocity changes by -1 (because 3 < 5). 
    # However, if the positions on a given axis are the same, 
    # the velocity on that axis does not change for that pair of moons.
    for (i,a) ∈ enumerate(copy(moons))
        for b ∈ copy(moons)
           if a != b 
                if a.px < b.px
                   moons[i].vx = moons[i].vx+1
                end
                if a.py < b.py
                   moons[i].vy = moons[i].vy+1
                end
                if a.pz < b.pz
                   moons[i].vz = moons[i].vz+1
                end   
                if a.px > b.px
                  moons[i].vx = moons[i].vx-1
                end
                if a.py > b.py
                   moons[i].vy = moons[i].vy-1
                end
                if a.pz > b.pz
                   moons[i].vz = moons[i].vz-1
                end             
            end
        end
    end
    #println(moons)
    #Once all gravity has been applied, 
    #apply velocity: simply add the velocity of each moon to its own position. 
    #For example, if Europa has a position of x=1, y=2, z=3 
    #and a velocity of x=-2, y=0,z=3, then its new position would be x=-1, y=2, z=6. 
    #This process does not modify the velocity of any moon.
    for a ∈ moons 
        a.px = a.px + a.vx
        a.py = a.py + a.vy
        a.pz = a.pz + a.vz
    end
    moons
end

io = Moon(-7, -1, 6,     0,0,0)
europa = Moon(6, -9, -9,      0,0,0)
ganymede = Moon(-12, 2, -7,   0,0,0)
callisto = Moon(4, -17, -12, 0,0,0)

moons = [io,europa,ganymede,callisto]
dowork(moons)
````

````julia
function energy(moon)
    (abs(moon.px) + abs(moon.py) + abs(moon.pz)) * (abs(moon.vx) + abs(moon.vy) + abs(moon.vz))
end;
````
